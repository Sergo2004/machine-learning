```python

# Импортируем необходимые библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Читаем данные из CSV файла
df = pd.read_csv('Coffe_sales.csv')
```

```python
# Определяем типы данных
# Числовые переменные
numeric_features = ['hour_of_day', 'money', 'Weekdaysort', 'Monthsort']
```

```python
# Категориальные переменные
categorical_features = ['cash_type', 'coffee_name', 'Time_of_Day', 'Weekday', 
                       'Month_name', 'Date', 'Time']
```

```python
# Разделяем данные на числовые и категориальные
df_numeric = df[numeric_features]
df_categorical = df[categorical_features]
```

```python
# Визуализация числовых данных
plt.figure(figsize=(12, 8))
```

```python
# Гистограммы для числовых признаков
plt.subplot(2, 2, 1)
sns.histplot(df_numeric['hour_of_day'], bins=24, kde=True)
plt.title('Распределение часов продаж')
```

```python
plt.subplot(2, 2, 2)
sns.histplot(df_numeric['money'], kde=True)
plt.title('Распределение стоимости')
```

```python
plt.subplot(2, 2, 3)
sns.boxplot(x=df_numeric['Weekdaysort'])
plt.title('Распределение по дням недели')
```

```python
plt.subplot(2, 2, 4)
sns.boxplot(x=df_numeric['Monthsort'])
plt.title('Распределение по месяцам')
```

```python
plt.tight_layout()
plt.show()
```

```python
# Визуализация категориальных данных
plt.figure(figsize=(14, 10))
```

```python
# Диаграммы для категориальных признаков
plt.subplot(2, 2, 1)
df_categorical['cash_type'].value_counts().plot(kind='bar')
plt.title('Распределение по типу оплаты')
```

```python
plt.subplot(2, 2, 2)
df_categorical['coffee_name'].value_counts().head(10).plot(kind='bar')
plt.title('Топ-10 популярных напитков')
```

```python
plt.subplot(2, 2, 3)
df_categorical['Time_of_Day'].value_counts().plot(kind='bar')
plt.title('Распределение по времени суток')
```

```python
plt.subplot(2, 2, 4)
df_categorical['Weekday'].value_counts().plot(kind='bar')
plt.title('Распределение по дням недели')
```

```python
plt.tight_layout()
plt.show()
```

```python
# Корреляционный анализ числовых признаков
plt.figure(figsize=(8, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Корреляция между числовыми признаками')
plt.show()
```
